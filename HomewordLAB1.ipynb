```python
# sau đó ta lấy các kiểu dữ liệu là Float để tính bằng cách 
dulieufloat = dulieuxettuyendaihoc.dtypes == float 
cotfloat = dulieuxettuyendaihoc.columns[dulieufloat].tolist() # lấy tất cả cột ra có chứa dữ liệu float, sau đó cho nó thành 1 cái list 
print(cotfloat) # và đây là kết quả, như vậy chúng ta đã lấy tất cả cột nào chứa dữ liệu float, cho vào list như hình bên dưới
```

```python
""" tiếp theo thay vì có 10 cột có chứa dữ liệu thiếu thì ta phải viết đúng 10 lần ( tương tự câu 4) 
.Làm như vậy sẽ mất thời gian, nên ta cho chạy vòng lập for, để nó tự thấy chỗ nào thiếu là nó tự điền là số 0 """
for col in cotfloat : 
  dulieuxettuyendaihoc[col].fillna(dulieuxettuyendaihoc[col].mean(),inplace = True)   
dulieuxettuyendaihoc.head()
```

```python
# Do đề bai kêu TBM6 là TMB3 nên mình sẽ đổi tên 
dulieuxettuyendaihoc.rename(columns={"TBM6":"TBM3"},inplace=True) 
dulieuxettuyendaihoc.head() # và dây là kết quả
```

```python
# Câu 8 Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp
# 10, 11, 12 như sau:
#  Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y)
#  Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB)
#  Từ 6.5 đến dưới 8.0: khá (kí hiệu là K)
#  Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G)
#  Từ 9.0 trở lên: xuất sắc (kí hiệu là XS) 
#CT dulieuxettuyendaihoc.loc[điều kiện, tên cột mới] = gán giá trị cho cột mới  
#CT dulieuxettuyendaihoc.loc[điều kiện và điều kiện, tên cột mới] = gán giá trị cho cột mới
```

```python
# Tham khảo: Trần Đức Toàn
```

```python
for i in ["1","2","3"] : 
  dulieuxettuyendaihoc.loc[dulieuxettuyendaihoc["TBM"+i] < 5.0,"XL"+i] = "Y" 
  dulieuxettuyendaihoc.loc[(dulieuxettuyendaihoc["TBM"+i] >= .0) & (dulieuxettuyendaihoc["TBM"+i] < 6.5),"XL"+i] ="TB"  
  dulieuxettuyendaihoc.loc[(dulieuxettuyendaihoc["TBM"+i] >=6.5) & (dulieuxettuyendaihoc["TBM"+i] < 8.0),"XL"+i] ="K" 
  dulieuxettuyendaihoc.loc[(dulieuxettuyendaihoc["TBM"+i] >= 8.0) & (dulieuxettuyendaihoc["TBM"+i] < 9.0),"XL"+i] ="G" 
  dulieuxettuyendaihoc.loc[(dulieuxettuyendaihoc["TBM"+i] >= 9.0) ,"XL"+i] ="XS"
dulieuxettuyendaihoc.head()
```
