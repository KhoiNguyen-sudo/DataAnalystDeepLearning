```python
import pandas as pd
import seaborn as sns
df_patientheartrate = pd.read_csv('https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/patient_heart_rate.csv')
print(df_patientheartrate)
```

```python
# thêm tên cột ( header) vào  df_patientheartrate 
ten_cot = ["ID", "Name", "Age", "Weight", "m0006", "m0612","m1218", "f006", "f0612", "f1218"] 
# Sau đó tiến hành gán  
#CT : ten biens = pd.read_csv("đường link dữ liệu",names = tencot)
df_patientheartrate = pd.read_csv('https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/patient_heart_rate.csv', names= ten_cot) 
print(df_patientheartrate)
```

```python
# Vấn đề 2: Xử lý vấn đề một cột lưu hỗn hợp nhiều dữ liệu, ở đây là cột ―Name‖ chứa bao 
# gồm ―Firstname‖ và ―Lastname‖, giải pháp là ta sẽ tách ra làm 2 cột 
```

```python
#CÁCH TÁCH CỘT DÙNG SPLIT 
# TÊN BIẾN [["tên cột mới thứ 1","tên cột mới thứ 2"]] = tên biên["tên cột muốn tách"].str.split(expand = True)
df_patientheartrate[['Firstname', 'Lastname']]=df_patientheartrate['Name'].str.split(expand = True)
df_patientheartrate.head(100)
```

```python
#Xóa cột Name .Dùng hàm drop
# CT: TÊN BIẾN = TÊN BIẾN.drop("TÊN CỘT MUỐN XÓA",AXIS=1) 
#AXIS: TÊN CỘT ĐÓ NẰM Ở CỘT THỨ MẤY
df_patientheartrate = df_patientheartrate.drop('Name',axis=1)
df_patientheartrate.head(100)
```

```python
# Vấn đề 3: Cột Weight có vấn đề về không thống nhất các đơn vị đo lường trong dữ liệu. 
# Ta sẽ chuyển các đơn vị về thành đơn vị chuẩn ―kg‖
```

```python
weight=df_patientheartrate['Weight']
for i in range(len(weight)):
  x = str(weight[i])
  if "lbs" in x[-3:]:
    #remove the lbs from the value
    x = x[:-3]
    float_x = float(x)
    y = int(float_x/2.205)
    y = str(y)+"kgs"
    weight[i] = y
print(df_patientheartrate.head(100))
```

```python
# Vấn đề 4: Vấn đề về xuất hiện dòng dữ liệu rỗng (không có giá trị: NaN). Giải pháp có 
#thể đưa ra là xóa bỏ
```

```python
#dùng hàm dropna để xóa dữ liệu rỗng
#CT tên biến.dropna(how="all",inplace=True)
df_patientheartrate.dropna(how="all", inplace=True)
df_patientheartrate.head(100)
```

```python
# Vấn đề 5: Có nhiều dòng dữ liệu bị trùng lắp thông tin hoàn toàn[fullname, lastname, 
# age, weight,....], giải pháp đưa ra là chỉ giữ lại một dòng dữ liệu, tuy nhiên giải pháp phải 
# dựa trên nghiệp vụ của tập dữ liệu và quan sát của người xử lý.
```

```python
# xóa những dữ liệu trùng dùng hàm drop_duplicates 
##tên biến  =tiến biến. drop_duplicates(subset=["tên cột có chứa dữ liệu trùng lần 1","tên cột có chứa dữ liệu trùng lần 2",...])
df_patientheartrate = df_patientheartrate.drop_duplicates(subset=['Firstname','Lastname', 'Age', 'Weight'])
df_patientheartrate.head(100) 
```

```python
# Vấn đề 6: Xuất hiện dữ liệu bị ảnh hưởng bởi lỗi non-ASCII, không định dạng ASCII. 
# Giải pháp: Tùy vào nghiệp vụ ta có thể: xóa dữ liệu tại đó, thay thế bằng dữ liệu khác 
# hoặc thay bằng việc đánh dấu bằng một kí tự khác (ví dụ: ‗warning‘)
```

```python
df_patientheartrate.Firstname.replace({r'[^\x00-\x7F]+':''}, regex=True,inplace=True)
df_patientheartrate.replace({r'[^\x00-\x7F]+':''}, regex=True,inplace=True)
df_patientheartrate.head(100)
```

```python
# Vấn đề 7: ―Missing values‖, vấn đề này xảy ra tại các cột ―Age‖, ―Weight‖ và ―Heart 
# Rate‖. Thiếu dữ liệu (dữ liệu không đầy đủ) là vấn đề xảy ra nhiều trong các nguồn dữ 
# liệu do nhiều nguyên nhân chủ quan lẫn khách quan.Có một vài giải pháp để xử lý vấn đề 
# này, chủ yếu dựa trên kinh nghiệm và nghiệp vụ về tập dữ liệu đó. Một số giải pháp đưa 
# đề xuất từ chuyên gia như sau: 
# a. Deletion: Remove records with missing values
# b. Dummy substitution: Replace missing values with a dummy but valid value: e.g.: 0 for numerical values.
# c. Mean substitution: Replace the missing values with the mean.
# d. Frequent substitution: Replace the missing values with the most frequent item.
# e. Improve the data collector: Your business folk will talk to the clients and inform them about why it is worth fixing the problem with the data collector.
# Yêu cầu:
# Thống kê thông tin dữ liệu thiếu trên từng biến Age và Weight
# Yêu cầu xử lý dữ liệu thiếu như sau: Nếu dòng nào có Age hoặc Weight có dữ liệu thì phần Age hoặc Weight được tính như bên dưới, nếu thiếu cả 2 thông tin thì xóa dòng o Age: Giá trị thay thế là mean của các giá trị trong cột Age o Weight: Giá trị thay thế là mean của các giá trị trong cột Weight theo nhóm giới tính
```

```python
index_ageweightnull = df_patientheartrate[df_patientheartrate['Age'].isna() & df_patientheartrate['Weight'].isna()].index
index_ageweightnull
```

```python
df_patientheartrate.drop(index_ageweightnull, inplace=True)
df_patientheartrate.head(100)
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
sns.heatmap(df_patientheartrate.isna(), 
            yticklabels=False, 
            cbar=True, 
            cmap='viridis')
```

```python
df_patientheartrate['Age'].fillna(df_patientheartrate['Age'].mean(), inplace =False)
```

```python
#Vấn đề 8
#"một cột chứa quá nhiều thông tin cần được phân rã", như trong bài toán này ta thấy header "m0006" chứa các nội dung bao gồm: m -> male, 1218 ~ 12-18 (mm-dd). Còn giá trị thì là kết quả huyết áp
```

```python
df = pd.melt(df_patientheartrate, id_vars=['ID', 'Age', 'Weight', 'Firstname', 'Lastname'], value_name="PulseRate",var_name="sex_and_time").sort_values(['ID', 'Age', 'Weight', 'Firstname', 'Lastname']) 
df
```

```python
#Extract sex, hour lower bound and hour upper dound group
tmp_df= df["sex_and_time"].str.extract("(\D)(\d+)(\d{2})", expand=True) 
tmp_df.head(100)
```

```python
#Name columns
tmp_df.columns=["Sex","hours_lower","hours_upper"] 
tmp_df.head(100)
```

```python
#Create Time column based on "hours_lower" and "hours_upper" columns
tmp_df["Time"]=tmp_df["hours_lower"] + "-"+ tmp_df["hours_upper"]

tmp_df.head(5)
```

```python
#Merge
df =pd.concat([df, tmp_df], axis=1) 
df.head(100)
```
